**Correlation Analysis**

In this section, we'll basically explore the relationship between various factors and prices. Since price is a crucial aspect for each car, it is important to examine the different variables that may be correlated with it. To accomplish this, we will consider two different approaches. the 1st approach involves analyzing the correlated numeric variables and price using 'Pearson correlation'. The 2nd approach will assess the correlation between categorical variables and prices through 'test'.

**Numerical Data (Pearson)**

```python
fig, axes = plt.subplots(2, 3, figsize=(25, 8))
axes = axes.flatten()

for i, col in enumerate(numeric_columns):
    sns.regplot(x=col,
                y='Price',
                data=df,
                ax=axes[i],
                scatter_kws={"color": "red"},
                line_kws={"color": "black"})
    
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('')

plt.tight_layout()
plt.show()
```

**Categorical Data (ANOVA)**

For categorical variables, the 1st step is to determine whether a specific categorical variables has an impact on the price, meaning we need to assess it's significance. To do this we need to conduct an ANOVA test to compare the differences between means. After identifying the significant variables, we then perform pairwise descriptive analysis for the category within the relevant categorical variable, allowing for the comparison of their effects on the price(s).

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
for i in cat_columns:
    formula = 'Price ~ {}'.format(i)
    try:
        model = ols(formula, data=df).fit()
        anova = sm.stats.anova_lm(model, typ=2)
        p_value = anova.iloc[0, 3]
        print('P-value for Price - {}: {}'.format(i, p_value))
    except Exception as e:
        print(f'Error fitting ANOVA for {i}: {e}')
```

According to the expected output, against the returned values:

   - 'Fuel' and 'Transmission' are significant to explain the price of the car, due to the p_value is less than 0.05, and we reject null hypothesis...
   - However, 'Brand' and 'Model' might not bring positive impact on our model.

In this case, for the next step, we basically want to see for each label in the above categorical variables, to determine which of them has the most impact on the price.  Therefore there's need to create a function that calculates the mean for each label in every single categorical variable (just the 1st 5 labels with regards to strength to be specific!).

```python
def mean_pairwise(cat_var):
    """Return top 5 labels by mean Price for a categorical variable."""
    # accept either a column name (str) or a pandas Series/array-like
    try:
        if isinstance(cat_var, str):
            group_col = cat_var
        else:
            # if user passed an index from cat_columns (e.g. cat_columns[0]),
            # ensure we handle both name and Series inputs
            group_col = cat_var
        mean_by = df.groupby(group_col)['Price'].mean().dropna()
        mean_by = mean_by.sort_values(ascending=False).to_frame(name='Price')
        return mean_by.head(5)
    except KeyError:
        print(f'Column {cat_var} not found in dataframe')
        return pd.DataFrame(columns=['Price'])
    except Exception as e:
        print(f'Error computing mean_pairwise for {cat_var}: {e}')
        return pd.DataFrame(columns=['Price'])
```
